```python
#Let us try to get the the middle of the classes of the income range and on adding to secondary income we can get the total month income.
# First let us change the column by replacing the column with the individual coresspondin mean value
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('Below 5000',2500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('5000 – 10,000',7500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('11,000 – 20,000',15500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('21,000 – 50,000', 35500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('51,000 – 80,000', 65500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('81,000 – 100,000', 90500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('110,000 – 200,000',155000)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('201,000 – 300,000',250500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('301,000 – 500,000',400500)
df['What is your steady/constant income range in a month?'] = df['What is your steady/constant income range in a month?'].replace('500,000 – above',750000)
```

```python
# Let us Create a new column for total income
df['Total Monthly Income'] = df['What is your steady/constant income range in a month?']+ df['How much do you think you earn from all other sources of income in a month']    
TotalIncome= df.groupby('Total Monthly Income').mean()
TotalIncome
```

```python
# Let us Create a new column for net income
df['Net Income'] = df['What is your steady/constant income range in a month?'] - df['On an average, how much do you think you spend in a month?']    
df['Net Income'].value_counts()
# Project analysis show that there are defict monthly income.
```

```python
# Let us try to see if there is any correlation between expenses and Total Monthly Income
from scipy.stats import pearsonr
corr,_ = pearsonr(df['Total Monthly Income'], df['On an average, how much do you think you spend in a month?'])
print ('pearsons Correlation:%.3f' %corr)
```

Explorative Data analysis shows that these two variables  expenses and Total Monthly Income are positively correlated, which suggests a high leve of relationship between Total Monthly Income and Monthly Expenses.ie the expenses goes higher with higher monthly income.
